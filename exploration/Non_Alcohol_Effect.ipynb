```python
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
maths_data = pd.read_csv("../data/Maths.csv")
maths_data["G2-G1"] = maths_data["G2"] - maths_data["G1"]
maths_data["G3-G2"] = maths_data["G3"] - maths_data["G2"]
```

```python
maths_data.columns
```

## Absences vs grade

```python
def plot_box(factor, grade):
    alcohol_list = maths_data[factor].tolist()
    grade_list = maths_data[grade].tolist()

    box_dict = {}
    for i in range(len(alcohol_list)):
        if alcohol_list[i] in box_dict:
            box_dict[alcohol_list[i]].append(grade_list[i])
        else:
            box_dict[alcohol_list[i]] = []
    
    X = sorted(list(set(alcohol_list)))
    plt_data = []
    for x in X:
        plt_data.append(box_dict[x])
    return plt_data


```

```python
d = plot_box("traveltime", "G3")
t = matplotlib.pyplot.boxplot(d)
```

```python
d = plot_box("studytime", "G3")
t = matplotlib.pyplot.boxplot(d)
```
