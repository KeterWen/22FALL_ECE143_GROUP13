# Data Overview

```python
import pandas as pd
import matplotlib
import seaborn as sns
```

## Maths.csv

```python
maths_data = pd.read_csv("../data/Maths.csv")
maths_data["G2-G1"] = maths_data["G2"] - maths_data["G1"]
maths_data["G3-G2"] = maths_data["G3"] - maths_data["G2"]
```

```python
subplot = sns.heatmap(maths_data.corr(numeric_only=True))
subplot.get_figure().savefig("../chart/heat_map.png")
```

```python
for column_name in maths_data.columns:
    print(column_name)
    subplt = maths_data[column_name].plot.box()
    subfig = subplt.get_figure()
    subfig.savefig(f"../chart/{column_name}_distribution.png")
```
