```python
'''The following is a heatmap that determines how correlated each of the features (variables) are too
determining a student's grades ranging from their gender, absence rate, to how much alchohol they drink
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #statist graph package
math_filepath = "Maths.csv"
portuguese_filepath = "Portuguese.csv"


# read the csv file to extract data
data = pd.read_csv('Maths.csv')

df = pd.DataFrame(data)
X = list(df.iloc[:, 1])
Y = list(df.iloc[:, 32])

#This creates a heat map to determin how related the different variables are
h_labels = [x.replace('_', ' ').title() for x in
            list(df.select_dtypes(include=['number', 'bool']).columns.values)]

fig, ax = plt.subplots(figsize=(10,6))
_ = sns.heatmap(df.corr(), annot=True, xticklabels=h_labels, yticklabels=h_labels, cmap=sns.cubehelix_palette(as_cmap=True), ax=ax)
```

```python
'''From this data we can determine that grades are somewhat correlated to age, studytime,
travel time, father's education
'''
'''Variables such as absences. health, freetime, drinking alchohol have near neglible impact'''
'''Grades are more highly correlated to previous failures and mothers education'''
```
