import numpy as np

data = pd.read_csv('Maths.csv')
numGrades = 3#3
numPredict = 30#29
df = pd.DataFrame(data)
for i in range(26,numPredict):
    for j in range(2,numGrades):

       # 'Python':35}

        plt.scatter(predict_dict.keys(),predict_dict.values())
        plt.xlabel(df.iloc[:, i].name)
        plt.ylabel("Grades "+ str(j+1))
        plt.title(" Average Grades Per " + str(df.iloc[:,i].name))
        plt.show()
        reg = linear_model.LinearRegression()
        if i == 27 or i == 28:
            xvar = np.array([s for s in predict_dict.keys()]).reshape(-1,1)
            yvar = np.array([s for s in predict_dict.values()])
