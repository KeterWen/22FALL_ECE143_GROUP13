# Alcohol Effect

```python
import matplotlib
```

```python
def plot_box(alcohol, grade):
    alcohol_list = maths_data[alcohol].tolist()
    grade_list = maths_data[grade].tolist()

    box_dict = {}
    for i in range(len(alcohol_list)):
        if alcohol_list[i] in box_dict:
            box_dict[alcohol_list[i]].append(grade_list[i])
        else:
            box_dict[alcohol_list[i]] = []
    
    X = sorted(list(set(alcohol_list)))
    plt_data = []
    for x in X:
        plt_data.append(box_dict[x])
    return plt_data


```

```python
d = plot_box("Walc", "G3")
t = matplotlib.pyplot.boxplot(d)
```

```python
d = plot_box("Walc", "G2")
t = matplotlib.pyplot.boxplot(d)
```
